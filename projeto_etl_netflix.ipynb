Tratamento e Limpeza (Transform) 

1.	Tratamento de Visualizações (Viewership):
    o	Remova o texto "M Streams" da coluna, deixando apenas os números.

    o	Converta a coluna para decimal (float).

    o	Renomeie a coluna para Viewership_Millions.

2.	Tratamento de Duração (Duration):
    o	O texto está no formato "S4; 730 min".

    o	Eu quero que você divida isso em duas colunas novas:
       
        	Uma coluna chamada Season contendo apenas a temporada (ex: "S4").
        
        	Uma coluna chamada Duration_Minutes contendo apenas os minutos em formato de número inteiro (ex: 730).


```python
# removendo o texto e deixando apenas os números
df['Viewership'] = df['Viewership'].str.extract(r'([\d\.]+)', expand=False) 
```

```python
df.head()
```

```python
# convertendo a coluna para o tipo float
df['Viewership'] = df['Viewership'].astype(float)
```

```python
# renomeando a coluna
df = df.rename(columns={'Viewership': 'Viewership_Millions'})
```

```python
# revomendo as temporadas da coluna "Duration" e inserindo em uma nova
df ['Season'] = df ['Duration'].str.extract(r'([^\;]+)')
```

```python
df.head()
```

```python
df['Duration_Minutes'] = df['Duration'].str.extract(r';\s*(\d+)\s*min') # adicionando uma nova coluna com apenas os números
```

```python
df['Duration_Minutes'] = df['Duration_Minutes'].astype(int) # convertendo eles para números inteiros
```

```python
df.info()
```

3.	Tratamento de Datas:
    o	Converta a coluna de data (ex: "17-Jun-16") para o formato oficial de data (datetime) usando pd.to_datetime().


```python
df ['Release Date'] = pd.to_datetime( df['Release Date'], format= '%d-%b-%y') # formatando as datas
```

4.	Padronização de Texto (Genre):
    o	Faça um .str.strip() na coluna Genre para remover os espaços em branco que vieram sobrando no final das palavras (como em "Comedy Drama ").


```python
df ['Genre'] = df['Genre'].str.strip()
```

```python
df = df.drop(columns=['Duration']) # removendo a coluna não formatada
```

Fase 3: Regras de Negócio e Análise (Insights)

Com os dados perfeitamente limpos numéricos, extraia para o Diretor essas 3 informações (pode usar groupby, sort_values, etc):
   
    1.	O Maior Sucesso: Quais são os Top 5 títulos com o maior número de visualizações (Viewership_Millions)?
    
    2.	Qualidade Geográfica: Qual é a média da nota do IMDb agrupada por País (Country)? Quem produz as séries mais bem avaliadas?
   
    3.	A Maratona: Qual é o título (ou títulos) que tem a maior duração total em minutos (Duration_Minutes) da nossa base?


```python
top_cinco_title = df.sort_values(by='Viewership_Millions', ascending=False) # ordenando os valores da coluna de forma decrescente
```

```python
df = top_cinco_title
```

```python
df[['Title', 'Viewership_Millions']].head(5) # exibindo as cinco primeiras linhas
```

```python
media_por_pais = df.groupby('Country')['IMDb'].mean() # usando o groupby para agrupar os valores de cada país para depois tirar a média com o mean()
```

```python
print(media_por_pais)
```

```python
pais_campeao = media_por_pais.idxmax() # retorna o INDICIE do valor mais alto
```

```python
mais_avaliado = media_por_pais.max() # retorna o VALOR mais alto
```

```python
print(f'O país mais bem avaliado é: {pais_campeao} com {mais_avaliado} de IMDb')
```

```python
top_title_duration = df.sort_values(by='Duration_Minutes', ascending=False)[['Title', 'Duration_Minutes']].head(1) # agrupar os valores e deixando de forma decrescente e depois exibindo a primeira linha com o head(1)

top_title_duration
```
